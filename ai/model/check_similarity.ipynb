```python
import numpy as np
import random

# train dataset 집어넣기
dataset = []

B_forehand = np.load('../variable_dataset/Federer_B_Forehand_dataset.npy')
B_backhand = np.load('../variable_dataset/Federer_B_Backhand_dataset.npy')
B_backslice = np.load('../variable_dataset/Federer_B_BackSlice_dataset.npy')
B_forevolley = np.load('../variable_dataset/Federer_B_ForeVolley_dataset.npy')
B_backvolley = np.load('../variable_dataset/Federer_B_BackVolley_dataset.npy')
B_smash = np.load('../variable_dataset/Federer_B_Smash_dataset.npy')
B_serve = np.load('../variable_dataset/Federer_B_Serve_dataset.npy')

for i in B_forehand:
    dataset.append(i)
for i in B_backhand:
    dataset.append(i)
for i in B_backslice:
    dataset.append(i)
for i in B_forevolley:
    dataset.append(i)
for i in B_backvolley:
    dataset.append(i)
for i in B_smash:
    dataset.append(i)
for i in B_serve:
    dataset.append(i)

dataset = np.array(dataset)
```

```python
import matplotlib.pyplot as plt
from tslearn.clustering import TimeSeriesKMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

# 클러스터 수를 2에서 10까지 시도하여 최적의 클러스터 수 찾기
sil_scores = []
db_scores = []

for k in range(2, 20):
    model = TimeSeriesKMeans(n_clusters=k, metric="dtw", random_state=1234)
    labels = model.fit_predict(dataset)

    # 평가 지표 계산 (Silhouette Index와 Davies-Bouldin Index)
    sil_score = silhouette_score(dataset.reshape(dataset.shape[0], -1), labels, metric='euclidean')
    db_score = davies_bouldin_score(dataset.reshape(dataset.shape[0], -1), labels)

    sil_scores.append(sil_score)
    db_scores.append(db_score)

    print(k)

# 결과 시각화
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(2, 20), sil_scores, marker='o')
plt.title('Silhouette Index')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')

plt.subplot(1, 2, 2)
plt.plot(range(2, 20), db_scores, marker='o')
plt.title('Davies-Bouldin Index')
plt.xlabel('Number of clusters')
plt.ylabel('DB Score')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tslearn.metrics import dtw
from scipy.spatial import procrustes
from sklearn.metrics.pairwise import cosine_similarity
from mpl_toolkits.mplot3d import Axes3D
import joblib

similarity_model = joblib.load('./similarity_model.joblib')
centroids = similarity_model.cluster_centers_
remove_padding_centroids = []

for centroid in centroids:
    remove_padding_centroids.append(np.array([i for i in centroid if i[0] != 0]))

def remove_probability(data):
    data_index = np.array([i for i in range(100) if i % 4 != 0])
    new_data = data[:, data_index]
    return new_data

# 두 형상의 3차원 좌표 데이터
shape1_3d = remove_probability(remove_padding_centroids[9])
shape2_3d = np.load("../video/test_federer.npy")
shape2_3d = shape2_3d[4]
shape2_3d = remove_probability(shape2_3d[:len(remove_padding_centroids[9])])
shape3_3d = np.load("../video/test_video.npy")
shape3_3d = shape3_3d[363]
shape3_3d = remove_probability(shape3_3d[:len(remove_padding_centroids[9])])

print(dtw(shape1_3d, shape2_3d) / 60)
print(dtw(shape1_3d, shape3_3d) / 60)

cos_similarities = cosine_similarity(shape1_3d, shape2_3d)
cos_sim_diagonal = np.diag(cos_similarities)
average_cos_similarity = np.mean(cos_sim_diagonal)
print(shape1_3d.shape, shape2_3d.shape)
print(cos_similarities.shape)
print((cos_similarities + 1) / 2)
print()
print(cos_sim_diagonal)
print()
print(average_cos_similarity)
print()
cos_similarities1 = cosine_similarity(shape1_3d, shape3_3d)
cos_sim_diagonal1 = np.diag(cos_similarities1)
average_cos_similarity1 = np.mean(cos_sim_diagonal1)
print((cos_similarities1 + 1) / 2)
print()
print(cos_sim_diagonal1)
print()
print(average_cos_similarity1)
print()

distances = np.sqrt(np.sum((shape1_3d - shape2_3d)**2, axis=1))
total_distance = np.sum(distances)

distances1 = np.sqrt(np.sum((shape1_3d - shape3_3d)**2, axis=1))
total_distance1 = np.sum(distances1)

print(total_distance)
print(total_distance1)

max_distance_per_point = np.sqrt(100)
max_total_distance = max_distance_per_point * 60
print(max_total_distance)

print((total_distance / max_total_distance) * 100)
print((total_distance1 / max_total_distance) * 100)

# Procrustes 분석 수행
m1_3d, m2_3d, disparity_3d = procrustes(shape1_3d, shape3_3d)

# Procrustes 거리를 백분율로 변환
disparity_percentage = disparity_3d * 100

# 유사도 계산 (100% - Procrustes 거리 백분율)
similarity_percentage = (1 - disparity_3d) * 100

print(f"Procrustes 거리 (백분율): {disparity_percentage:.2f}%")
print(f"형상의 유사도 (백분율): {similarity_percentage:.2f}%")

# 3차원 형상과 변환된 형상을 시각화
fig = plt.figure(figsize=(12, 6))

# 원래 형상
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(shape1_3d[:, 0], shape1_3d[:, 1], shape1_3d[:, 2], c='r', label='Shape 1')
ax1.scatter(shape2_3d[:, 0], shape2_3d[:, 1], shape2_3d[:, 2], c='b', label='Shape 2')
ax1.scatter(shape3_3d[:, 0], shape3_3d[:, 1], shape3_3d[:, 2], c='g', label='Shape 3')
ax1.set_title('Original Shapes')
ax1.legend()

# 변환된 형상
ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(m1_3d[:, 0], m1_3d[:, 1], m1_3d[:, 2], c='r', label='Transformed Shape 1')
ax2.scatter(m2_3d[:, 0], m2_3d[:, 1], m2_3d[:, 2], c='b', label='Transformed Shape 2')
ax2.set_title('Transformed Shapes')
ax2.legend()

plt.show()
```
